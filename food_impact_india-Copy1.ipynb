```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = '/home/bhavya/Downloads/datasets/food_impact_india.csv'
df = pd.read_csv(file_path)
```

```python
print('dataset size: ',df.shape)
print('5 sample rows:')
df.sample(5)
```

```python
df.drop_duplicates(inplace = True)
df['Common_Diseases'] = df['Common_Diseases'].fillna('No disease found')
```

```python
df.info()
#pd.isnull(df).sum() -> code to count no of null values
# it was found that there are no null values
```

```python
# finding out health score and age ranges in this dataset
max_health_score = df['Health_Score'].max()
min_health_score = df['Health_Score'].min()

max_age = df['Age'].max()
min_age = df['Age'].min()

print('in this data, \nthe maximum health score in this data is ',max_health_score)
print('the minimum health score in this data is ',min_health_score)
print("max age: ",max_age)
print("min age: ",min_age)
```

```python
df.to_csv('cleaned_data.csv', index=False)
# this CSV file was then uploaded to Tableau from where the data visualizations were downloaded.
```
